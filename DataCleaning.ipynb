```python
import pandas as pd

input = pd.read_csv("james_bond_data.csv").convert_dtypes()
input.head()
```

```python
new_names = {"Release":"release_date",
            "Movie": "movie_title",
            "Bond": "bond_actor",
            "Bond_Car_MFG": "car_manufacturer",
            "US_Gross": "income_usa",
            "World_Gross": "income_world",
            "Budget ($ 000s)": "movie_budget",
            "Film_Length": "film_length",
            "Avg_User_IMDB": "imdb",
            "Avg_User_Rtn_Tom": "rotten_tomatoes",
            "Martinis": "martinis_consumed",
            "Kills_Bond": "bond_kills"}

data = input.rename(columns=new_names)
data.head()
data.info()
```

```python
data.loc[data.isna().any(axis="columns")]
```

```python
data.at[10, 'imdb'] = 7.1
data.at[10, 'rotten_tomatoes'] = 6.8

data.loc[10]
data.head()
```

```python
def remove_dollar(column):
    return float((column.replace('$','').replace(',', '').replace('.','')))
def multiply_price(column):
    return float(column*1000)
def remove_mins(column):
    return int(column.replace(' mins', ''))

data['income_usa'] = data['income_usa'].apply(remove_dollar)
data['income_world'] = data['income_world'].apply(remove_dollar)
data['movie_budget'] = (data['movie_budget'].apply(remove_dollar)).apply(multiply_price)
data['film_length'] = data.film_length.apply(remove_mins)
```

```python
data['release_date'] = pd.to_datetime(data['release_date'], format='%B, %Y')

```

```python
data = data.assign(release_year = lambda data: (data['release_date'].dt.year.astype('Int64')))
data.head()
```

```python
data['bond_actor'].value_counts()
data['bond_actor'] = data['bond_actor'].str.replace('Shawn Connery', 'Sean Connery').str.replace('Roger MOORE', 'Roger Moore')
```

```python
data['car_manufacturer'].value_counts()
```

```python
data['car_manufacturer'] = data['car_manufacturer'].str.replace('Astin Martin', 'Aston Martin')
```

```python
data[['film_length', 'martinis_consumed']].describe()
```

```python
data['film_length'] = data['film_length'].replace(1200, 120)
data['martinis_consumed'] = data['martinis_consumed'].replace(-6, 6)
```

```python
data[['film_length', 'martinis_consumed']].describe()
```
